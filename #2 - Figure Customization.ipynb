### • If we want to create 4 axes at the same time:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,6)
```

```python
y = np.arange(2,11,2)
```

```python
figure, axes = plt.subplots(nrows = 2, ncols = 2) 
# If you press Shift+Tab here, you will see "nrows = 1" and "ncols = 1" parameters as default. 
#This means, a 1x1 axes will be created as default and with this we can create 1 axes, but we want to create 4 graph so we need 2x2 axes. That's why we gave "nrows = 2" and "ncols = 2"
# 2x2 axes means, consider an area with 4 windows. Each window will be 1 axes.
```

### • As you can see, They're almost stuck with each other. We can show them better with "plt.tight_layout()" code:

```python
figure, axes = plt.subplots(nrows = 2, ncols = 2) 
plt.tight_layout()
```

***
### Let's create 2 axes instead of 4 and check their type and real form:

#### Type:

```python
figure, axes = plt.subplots(nrows = 2, ncols = 1) 
print(type(axes)) # Check output
plt.tight_layout()
```

#### Real form:

```python
figure, axes = plt.subplots(nrows = 2, ncols = 1) 
print(axes) # Check output
plt.tight_layout()
```

### • We have 4 axes. We can also see them one by one by their indexes:

```python
figure, axes = plt.subplots(nrows = 2, ncols = 1) 
print(axes[0])   # Printing axes[0], check the output
plt.tight_layout()
```

### • Since they are in list form, we can navigate on them and create lines for both:

```python
figure, axes = plt.subplots(nrows = 2, ncols = 1) 

for ax in axes:
    ax.plot(x, y) # We are drawing them a line
    
plt.tight_layout()
```

### • If you don't want to make them same:

```python
figure, axes = plt.subplots(nrows = 2, ncols = 1) 

axes[0].plot(x, y)
axes[0].set_title("First Axes") #We can name them too

axes[1].plot(x, x**3)
axes[1].set_title("Second Axes") # Naming the second axes

plt.tight_layout()
```

### Sometimes we want a figure to cover all window or cover a little size. To do this, we can decide figure size:

```python
figure = plt.figure(figsize = (10,10)) 
# If you press Shift+Tab you can see "figsize" parameter. It takes "x" and "y" as values. 
#This changes the figure window's size, you can understand it better on a file

axes = figure.add_axes([0, 0, 1, 1]) # Giving these values makes it cover all over the window
axes.plot(x, y, color = "magenta")
```

#### ► Example:

```python
figure = plt.figure(figsize = (3,6))

axes = figure.add_axes([0, 0, 1, 1])
axes.plot(x, y, color = "magenta")
```

## Operations on figures:

#### • Saving a figure to a file:

```python
figure,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,6))
axes[0].plot(x,y, color = "blue" )
axes[0].set_title("First Axes") # We give name to the first axes

axes[1].plot(x, x ** 0.5, color = "green")
axes[1].set_title("Second Axes")# We give name to the second axes

plt.tight_layout()
```

```python
figure.savefig("Figure 1.png") # The file has been created into the same folder as your file that you write this code
```

```python
figure.savefig("Figure 1.pdf") # We can save it as pdf file. With this way you can print it from a printer
```

#### • Drawing 3 lines on a single graphic:

```python
figure = plt.figure(figsize = (6,4))

axes = figure.add_axes([0, 0, 1, 1])
axes.plot(x, x ** 0.5)
axes.plot(x, x ** 2)
axes.plot(x, x ** 3)
```

• But here, it's hard to see which one is what line. To prevent this, we can name them:

```python
figure = plt.figure(figsize = (6,4))

axes = figure.add_axes([0, 0, 1, 1])
axes.plot(x, x ** 0.5, color = "red", label = "X Square Root") # We are giving names to them with "label" parameter
axes.plot(x, x ** 2, color = "#FAFA00", label = "X Square") # You can use hexadecimal color codes too, you may search it on google 
axes.plot(x, x ** 3, color = "#008CFA", label = "X Cube")

#In order to apply them you have to use the code below:
axes.legend()
```
