```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1, 6)
```

```python
y = np.arange(2, 11, 2)
```

### • We can change the line width with "linewidth" parameter. You can also use it as "lw" : 

#### ► Normal:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", linewidth = 4) # Changing the line width
```

#### ► Short:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", lw = 6) # We can use it like this too
```

***
### • We can change the line style with "linestyle" parameter. You can also use it as "ls" :

#### ► Normal:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", linewidth = 3, linestyle = "--") # Changing the line style
```

#### ► Short:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", linewidth = 3, ls = "-.") # We can also use it like this 
```

### • There are multiple ways to change the style:
 **'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'**

***
### • We can add markers with "marker" parameter:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", lw = 3, marker = "o") # Adding markers
```

• x has (1,2,3,4,5) values that's why those points are marked.

***
### • We can change marker size with "markezsize" parameter:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", lw = 3, marker = "o", markersize = 12) # Changing marker size
```

***
### • We can change marker color with "markerfacecolor" parameter:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", lw = 3, marker = "o", markersize = 12, markerfacecolor = "black") # Changing marker color
```

***
### • We can change marker edge color with "markeredgecolor" parameter:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", lw = 3, marker = "o", markersize = 12, markerfacecolor = "black", markeredgecolor = "blue") #Changing marker edge color
```

***
### • We can change marker edge width with "markededgewidth" parameter:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", lw = 3, marker = "o", markersize = 12, markerfacecolor = "black", markeredgewidth = 3) # Red edges will be wider
```

***
### • We can change values of graphics:

#### ► Changing x axis:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", lw = 3, marker = "o", markersize = 12, markerfacecolor = "black", markeredgewidth = 3)

axes.set_xlim(0,10) # The line will stop at 5, because we gave x axis until 5 but graphic will continue until 10
```

#### Changing both axes:

```python
figure = plt.figure()

axes = figure.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, color = "red", lw = 3, marker = "o", markersize = 12, markerfacecolor = "black", markeredgewidth = 3)

axes.set_xlim(0,10) # The line will stop at 5 because we gave x axis until 5, but graphic will continue until 10
axes.set_ylim(0,50) # The line will stop at 50 because we gave y axis as x axis' square (which means the last value is 5**2), but graphic will continue until 50
```
